```python
import pandas as pd
```

```python
df = pd.read_csv(r"./archive/ks-projects-201801.csv")
df.head()
```

```python
df_colum_unique_count = pd.DataFrame()
df_colum_unique_value = pd.DataFrame()
for colum in df.columns:
    unique_value_count = len(df[colum].unique())
    df_colum_unique_count[colum] = [unique_value_count]
    df_colum_unique_value[colum] = pd.DataFrame(df[colum].unique())

df_colum_unique = pd.concat([df_colum_unique_count, df_colum_unique_value])
df_colum_unique
```
